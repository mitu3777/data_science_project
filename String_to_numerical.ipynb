```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\shree\OneDrive\Documents\my excel\Bmi_male_female.csv")
df.head()
```

```python
x=df[["Height","Weight"]]
y=df.Gender
```

```python
Y=pd.get_dummies(y)
x
```

```python
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x,Y)
```

```python
ans = model.predict([[130,90]])
```

```python
if ans[0][0]==True:
    print ("Female")
else:
    print("Male")
```
