```python
import matplotlib.pyplot as plt
```

```python
# Drop rows where 'area' or 'price' is missing
filtered_data = df.dropna(subset=["area", "price"])

# Assign x and y from the cleaned data
x = filtered_data[["area"]]
y = filtered_data["price"]

plt.title("area vs price")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(x,y,color="red",s=100)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
x = filtered_data[["area", "bedroom", "age"]]  # use all 3 columns
y = filtered_data["price"]

model = LinearRegression()
model.fit(x, y)

# Now predict using all 3 inputs
model.predict([[5000, 4, 3]])
```
